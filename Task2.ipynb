### Task 2 : Prediction using Decision Tree
#### Aim : To create the decision tree classifier and to visualize it graphically.

```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as datasets
```

```python
# Loading the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

Here, iris_df contains 4 columns as shown above.
Hence Target column is fetched the 'iris' dataset and not from iris_df

```python
y = iris.target
print(y)
```

```python
print('The number of kinds of iris flower is' ,np.unique(y))
```

```python
print(iris_df.info())
```

```python
# Statistical Information

iris_df.describe()
```

```python
iris_df = pd.DataFrame(iris_df)
iris_df.head()
```

```python
sns.pairplot(iris_df, kind = 'scatter',  diag_kind = 'kde', height = 1.8);   # hist - histogram
```

 #### Note : criterion - "gini” for the Gini impurity and “entropy” for the information gain.

```python
# Fitting the data and visualizing the graph.

from sklearn.tree import DecisionTreeClassifier
dtr = DecisionTreeClassifier(criterion = 'entropy')
dtr = dtr.fit(iris_df,y)

from sklearn import tree
plt.figure(figsize=(30,18))
tree.plot_tree(dtr, feature_names = iris_df.columns, 
               precision = 2, rounded = True, filled = True);
```
